```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\aswin\Downloads\world_population.csv')
df.head()
```

```python
# Filtering using Comparison Operators

df[df['Rank'] <= 10]
```

```python
# To check if the the countries listed in the list are present in the DataFrame using isin()

countries = ['India', 'Mexico']

df[df['Country'].isin(countries)]
```

```python
# To check if the the countries listed in the list are present in the DataFrame using isin()

df[df['Country'].str.contains('States')]
```

```python
# Filtering using indexes by rows

# Setting the Country as an index

df2 = df.set_index('Country')
df2.head()
```

```python
# Filter - Default filtering axis is by column

df2.filter(items = ['Continent' , 'Capital'])
df2.head()
```

```python
# Filtering by rows

df2.filter(items = ['Kenya'] , axis = 0)
```

```python
# Filtering using like

df2.filter(like = 'Au' , axis = 0)
```

```python
# Using loc

df2.loc['Belgium']
```

```python
# Using iloc

df2.iloc[:5]
```

```python
# Ordering or Sorting values

df3 = df2

df3 = df3[df3['Rank'] < 10]
df3
```

```python
df3.sort_values('Rank')
```

```python
# sorting by descending

df3.sort_values('Rank' , ascending = False)
```

```python
# sorting by multiple columns

df3.sort_values(['Continent', 'Country'], ascending = [True, True])
```
